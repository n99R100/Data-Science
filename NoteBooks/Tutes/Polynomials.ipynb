# Roots

```python
r = np.poly1d([1,2])
r
```

```python
print(np.poly1d(r))
```

```python
np.roots(r)
```

```python
r = np.poly1d([3,2,0])
r
```

```python
print(np.poly1d(r))
```

```python
np.roots(r)
```

```python
r = np.poly1d([3,2,-6])
r
```

```python
print(np.poly1d(r))
```

```python
np.roots(r)
```

```python
print(np.poly1d(np.poly1d([4,6], True)))
```

<h1><font face = "Optima"; size = 20; color = "red">Radicands</fonts></h1>

```python
x = sym.Symbol('x')
x
```

```python
sym.sqrt(x)
```

```python
x**sym.Rational(2,3)
```

```python
sym.sqrt(x+2)
```

```python
(x+2)**sym.Rational(2,3)
```

```python
x,y = sym.symbols('x,y')
```

```python
p = sym.sqrt(2*x)*sym.sqrt(3*y)
p
```

```python
p.simplify()
```

```python
sym.sqrt(2*x)/sym.sqrt(6*y)
```

```python
p = -1*((8*x**6)/125)**sym.Rational(1,3)
p
```

```python
p.evalf()
```

```python
a,b,c,m,n = sym.symbols('a,b,c,m,n')
```

```python
p = (sym.sqrt(m*n**3)/(a**2*sym.sqrt(c**(-3))))*((a**(-7)*n**(-2))/(sym.sqrt(m**2*c**4)))
p
```

```python
p.evalf()
```

```python
p = (sym.sqrt(250*x**2)/(sym.sqrt(100*b**3)))*((7*sym.sqrt(b))/sym.sqrt(125*x))
p
```

```python
p.evalf()
```
