```python
def f(x):
    return x**3

x_values = np.linspace(-5, 5, 1000)

df_dx = np.gradient(f(x_values), x_values)

func = go.Scatter(x = x_values, y = f(x_values), name = "$f(x)$")
der = go.Scatter( x = x_values, y = df_dx, name = '$f\'(x)$')

fig = go.Figure(data = [func,der])

fig.update_layout(
    title="$f \mapsto x: x^3, f\'(x)$",
    xaxis_title="x",
    yaxis_title="y",
)

fig.show()
```

```python
def f(x):
    return (np.sqrt(x) + 1/np.sqrt(x))**2

x_values = np.linspace(0.01, 2, 1000)

df_dx = np.gradient(f(x_values), x_values)

fig = go.Figure()

fig.add_trace(go.Scatter(x=x_values, y=[f(x) for x in x_values], mode='lines', name='f(x) = (sqrt(x) + 1/sqrt(x))^2'))
fig.add_trace(go.Scatter(x=x_values, y=df_dx, mode='lines', name="f'(x)"))

fig.update_layout(
    title="Plot of f(x) and its Derivative",
    xaxis_title="x",
    yaxis_title="y",
)

fig.show()
```

```python
def f(x):
    return (24 * x**7 - 9 * x**4) / (3 * x**3)

x_values = np.linspace(-5, 5, 1000)

f_values = [f(x) for x in x_values]

f_prime_values = np.gradient(f_values, x_values)

f_double_prime_values = np.gradient(f_prime_values, x_values)

fig = go.Figure()

fig.add_trace(go.Scatter(x=x_values, y=f_values, mode='lines', name='f(x)'))
fig.add_trace(go.Scatter(x=x_values, y=f_prime_values, mode='lines', name="f'(x)"))
fig.add_trace(go.Scatter(x=x_values, y=f_double_prime_values, mode='lines', name="f''(x)"))

fig.update_layout(
    title="Plot of f(x), f'(x), and f''(x)",
    xaxis_title="x",
    yaxis_title="y",
)

fig.show()
```

## Area under the curve

## $y = x(6-x)$

```python
f = x*(6-x)
f
```

```python
def fi(x):
    return x * (6 - x)
```

```python
x_values = np.linspace(-1, 7, 1000)
y_values = fi(x_values)
x2 = np.linspace(0,6,1000)
y2 = fi(x2)

fig = go.Figure()

fig.add_trace(go.Scatter(x=x_values, y=y_values, mode='lines', name='f(x) = x(6-x)'))

fig.add_trace(go.Scatter(x=x2, y=y2, fill='toself', fillcolor='rgba(0,100,80,0.2)', name='Shaded Area'))

fig.update_layout(
    title="Plot of f(x) = x(6-x) with Shaded Area",
    xaxis_title="x",
    yaxis_title="y",
)

fig.show()
```
