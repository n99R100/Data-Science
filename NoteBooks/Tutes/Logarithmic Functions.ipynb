```python
import numpy as np
```

```python
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
```

<h1> <font face = 'Times'; color = 'red'; size = 15> Logarithmic Functions

# 1. Common Logarithms

```python
np.log10([x for x in range(1,11)])
```

## 1.1 $y = \log(x)$ Plot

```python
x = np.linspace(0.01, 10, 100)
y1 = np.log10(x)
y2 = np.log10(x)

fig = make_subplots(rows=1, cols=2, subplot_titles=('$y \mapsto x: \log_{10}(x)$', '$y \mapsto x: log_{10}(x) ~in~ log ~scale$'))

trace1 = go.Scatter(x=x, y=y1, mode='lines', name='y = log10(x)')
trace2 = go.Scatter(x=x, y=y2, mode='lines', name='y = log10(x)')

fig.add_trace(trace1, row=1, col=1)
fig.add_trace(trace2, row=1, col=2)

fig.update_layout(
    title='$f \mapsto x: \log_{10}(x)$',
    yaxis=dict(title='$y \mapsto x:$'),
)

fig.update_xaxes(title='$x$', row=1, col=1)
fig.update_xaxes(type='log', title='$\log_{10}(x)$', row=1, col=2)

fig.show()
```

# 2. Natural Logarithms

## 2.1 $y = \ln(x)$ Plot

```python
x = np.linspace(0.01, 10, 100)
y1 = np.log(x)
y2 = np.log(x)

fig = make_subplots(rows=1, cols=2, subplot_titles=('$y \mapsto x: \ln(x)$', '$y \mapsto x: \ln(x) ~in~ log ~scale$'))

trace1 = go.Scatter(x=x, y=y1, mode='lines', name='y = ln(x)')
trace2 = go.Scatter(x=x, y=y2, mode='lines', name='y = ln(x)')

fig.add_trace(trace1, row=1, col=1)
fig.add_trace(trace2, row=1, col=2)

fig.update_layout(
    title='$f \mapsto x: \ln(x)$',
    yaxis=dict(title='$y \mapsto x:$'),
)

fig.update_xaxes(title='$x$', row=1, col=1)
fig.update_xaxes(type='log', title='$\ln(x)$', row=1, col=2)

fig.show()
```

# 3. Transformations of Logarithmic Functions

## 3.1 Horizontal shift $\left[ y \mapsto x: log(x \pm c) \right]$

```python
x = np.linspace(0.01, 10, 100)
y1 = np.log(x)
y2 = np.log(x + 1)
y3 = np.log(x + 2)
y4 = np.log(x)
y5 = np.log(x - 1)
y6 = np.log(x - 2)

fig = make_subplots(rows=1, cols=2, subplot_titles=('y = log(x+c)', 'y = log(x-c)'))

trace1 = go.Scatter(x=x, y=y1, mode='lines', name='y = log(x)')
trace2 = go.Scatter(x=x, y=y2, mode='lines', name='y = log(x+1)')
trace3 = go.Scatter(x=x, y=y3, mode='lines', name='y = log(x+2)')

fig.add_trace(trace1, row=1, col=1)
fig.add_trace(trace2, row=1, col=1)
fig.add_trace(trace3, row=1, col=1)

trace4 = go.Scatter(x=x, y=y4, mode='lines', name='y = log(x)')
trace5 = go.Scatter(x=x, y=y5, mode='lines', name='y = log(x-1)')
trace6 = go.Scatter(x=x, y=y6, mode='lines', name='y = log(x-2)')

fig.add_trace(trace4, row=1, col=2)
fig.add_trace(trace5, row=1, col=2)
fig.add_trace(trace6, row=1, col=2)

fig.update_layout(
    title='$y \mapsto x: \log(x \pm c)$',
    xaxis=dict(title='$x$'),
    yaxis=dict(title='$y$'),
)

fig.show()
```

## 3.2 Vertical shift $\left[ y \mapsto x: \log(x) \pm d \right]$

```python
x = np.linspace(0.01, 10, 100)
y1 = np.log(x)
y2 = np.log(x) + 1
y3 = np.log(x) + 2
y4 = np.log(x)
y5 = np.log(x) - 1
y6 = np.log(x) - 2

fig = make_subplots(rows=1, cols=2, subplot_titles=('y = log(x+d)', 'y = log(x-d)'))

trace1 = go.Scatter(x=x, y=y1, mode='lines', name='y = log(x)')
trace2 = go.Scatter(x=x, y=y2, mode='lines', name='y = log(x)+1')
trace3 = go.Scatter(x=x, y=y3, mode='lines', name='y = log(x)+2')

fig.add_trace(trace1, row=1, col=1)
fig.add_trace(trace2, row=1, col=1)
fig.add_trace(trace3, row=1, col=1)

trace4 = go.Scatter(x=x, y=y4, mode='lines', name='y = log(x)')
trace5 = go.Scatter(x=x, y=y5, mode='lines', name='y = log(x)-1')
trace6 = go.Scatter(x=x, y=y6, mode='lines', name='y = log(x)-2')

fig.add_trace(trace4, row=1, col=2)
fig.add_trace(trace5, row=1, col=2)
fig.add_trace(trace6, row=1, col=2)

fig.update_layout(
    title='$f \longmapsto x: \log(x) \pm d$',
    xaxis=dict(title='x'),
    yaxis=dict(title='y'),
)

fig.show()
```

## 3.3 Stretch and compress  $\left[ y \longmapsto x: \pm a \log(x) \right]$

```python
x = np.linspace(0.01, 10, 100)
y1 = np.log(x)
y2 = 2 * np.log(x)
y3 = -2 * np.log(x)
y4 = np.log(x)
y5 = 0.5 * np.log(x)
y6 = -0.5 * np.log(x)

fig = make_subplots(rows=1, cols=2, subplot_titles=('$y = \pm a \log(x)$', '$y = \pm (1/a) \log(x)$'))

trace1 = go.Scatter(x=x, y=y1, mode='lines', name='y = log(x)')
trace2 = go.Scatter(x=x, y=y2, mode='lines', name='2log(x)')
trace3 = go.Scatter(x=x, y=y3, mode='lines', name='-2log(x)')

fig.add_trace(trace1, row=1, col=1)
fig.add_trace(trace2, row=1, col=1)
fig.add_trace(trace3, row=1, col=1)

trace4 = go.Scatter(x=x, y=y4, mode='lines', name='y = log(x)')
trace5 = go.Scatter(x=x, y=y5, mode='lines', name='0.5log(x)')
trace6 = go.Scatter(x=x, y=y6, mode='lines', name='-0.5log(x)')

fig.add_trace(trace4, row=1, col=2)
fig.add_trace(trace5, row=1, col=2)
fig.add_trace(trace6, row=1, col=2)

fig.update_layout(
    title='$y \longmapsto x: \pm a \log(x)$',
    xaxis=dict(title='x'),
    yaxis=dict(title='y'),
)

fig.show()
```

## 3.4 $y \longmapsto x: \pm a \log(+bx \pm c) \pm d$

```python
x = np.linspace(0.01, 20, 100)
y = 2 * np.log(0.5 * x + 4) + 6

trace = go.Scatter(x=x, y=y, mode='lines', name='y = 2*log(0.5x+4)+6')

layout = go.Layout(
    title='$y = 2 \log(0.5x+4)+6$',
    xaxis=dict(title='$x$', range = (0,20)),
    yaxis=dict(title='$y$'),
    height = 500,
    width = 1000,
)

fig = go.Figure(data=[trace], layout=layout)
fig.show()
```
