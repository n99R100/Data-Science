# 1. Sequences

## 1.1 Arithmetic Sequences

<div class="alert alert-block alert-info">
<b><font color = 'blue'; size = 4; face = "Arial"> Definition:</font></b> <hr>
<font color = 'black'; size = 3; face = "Times">In an arithmetic sequence, each term is obtained by adding a fixed constant (called the <i>common difference</i>) to the previous term. The general form of an arithmetic sequence is $a_n = a_1 + d(n -1)$. Where, $a_1$ is the first term, and $d$ is the common difference.
</div>

- $\{-3, 1,5,9,13,..\}$

```python
seq = [-3,1,3,5,9,13]
```

```python
x = list(range(1,10))
y = [(-3)+(4)*(i-1) for i in range(1,10)]
```

```python
trace = go.Scatter(x = x, y = y, name = "Sequence", mode = "markers", marker = dict(size = 10))

fig = go.Figure(data = [trace], layout = dict(title = "Arithmatic Sequence", 
                                              xaxis = dict(title = "Term"), yaxis = dict(title = "Value"),
                                             height = 600, width = 800))

fig.show()
```

## 1.2 Geometric Sequences

- $\{1,2,4,8,16,...,\}$

```python
x = list(range(1,10))
y = [1*(4/2)**(i-1) for i in range(1,10)]
```

```python
trace = go.Scatter(x = x, y = y, name = "Sequence", mode = "markers", marker = dict(size = 10, color = "red"))

fig = go.Figure(data = [trace], layout = dict(title = "Geometric Sequence", 
                                              xaxis = dict(title = "Term"), yaxis = dict(title = "Value"),
                                             height = 600, width = 800))

fig.show()
```

## 1.3 Harmonic Sequences

- $\{ 1/5, 1/9, 1/13, 1/14 .....\}$

# 2. Series

<div class="alert alert-block alert-info">
    <b><font color = 'blue'; size = 4> Definition:</font></b> <hr>
A series is a mathematical concept that represents the sum of the terms of a sequence. In mathematics, a sequence is an ordered list of numbers or other mathematical objects, and a series is formed by adding up all the terms in this sequence.
</div>

## 2.1 Arithmetic Series

$$f \longmapsto i: 5 + (i -1) \cdot 3$$

```python
s = 0

for i in range(1,11):
    s += (5+(i-1)*3)

s
```

```python
s = (10*(5+32))/2
s
```

```python
a1 = 5
d = 3
n_terms = 10

arithmetic_series = [(i*(a1 + (a1 + (i - 1) * d)))/2 for i in range(1, n_terms + 1)]

trace = go.Scatter(x=list(range(1, n_terms + 1)), y=arithmetic_series, mode='lines+markers', marker = dict(size = 10), name='Arithmetic Series')

layout = go.Layout(
    title='Arithmetic Series',
    xaxis=dict(title='Sum to Term Number'),
    yaxis=dict(title='Sum'),
    height = 600,
    width = 800
)

fig = go.Figure(data=[trace], layout=layout)

fig.show()
```

## 2.2 Geometric Series

- $\{4,8,16,32,64,\}$

```python
s = (4*(1-(-2)**10))/(1-(-2))
s
```

```python
a1 = 4
r = -2
n_terms = 10

arithmetic_series = [(a1*(1-r**i))/(1-r) for i in range(1, n_terms + 1)]

trace = go.Scatter(x=list(range(1, n_terms + 1)), y=arithmetic_series, mode='lines+markers', marker = dict(size = 10), name='Geometric Series')

layout = go.Layout(
    title='Geometric Series',
    xaxis=dict(title='Sum to Term Number'),
    yaxis=dict(title='Sum'),
    height = 600,
    width = 800
)

fig = go.Figure(data=[trace], layout=layout)

fig.show()
```
