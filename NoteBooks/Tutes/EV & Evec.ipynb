<h1> <font size = 15; color = "Orange"; face = "Times"> Eigen Values and Eigen Vectors

- **Eigenvalues (Î»):**

An eigenvalue of a *square matrix A* is a **scalar value $\lambda$** such that when the matrix A is multiplied by a vector $v$, the result is a scaled version of the vector $v$. In other words, $A\cdot v = \lambda \cdot v$, where $v$ is the eigenvector corresponding to eigenvalue $\lambda$. The eigenvalue  $\lambda$ represents the scaling factor by which the eigenvector is stretched or compressed.

- **Eigenvectors (v):** 

An eigenvector of a square matrix A is a non-zero vector $v$ such that the matrix A, when multiplied by $v$, results in a scalar multiple of $v$. Mathematically, $A\cdot v = \lambda \cdot v$, where $v$ is the eigenvector and $\lambda$ is the corresponding eigenvalue.

```python
A = np.array([[3, 1],
              [2, 2]])
```

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
```
