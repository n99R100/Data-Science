```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
# Known data points
x_known = np.array([1,3,4,5,6,7,8,9,10])
y_known = np.array([9,4,5,7,4,4,5,8,5])
```

```python
# Interpolation function
interp_func = interp1d(x_known, y_known, kind='linear')
```

```python
# New points for interpolation and extrapolation
x_new = 1+np.random.rand(1,10)*5
x_new
```

```python
# Calculate interpolated values
y_interp = interp_func(x_new)
```

```python
# Plotting
#plt.plot(x_known, y_known, 'o', label='Known Data')
plt.plot(x_known, y_known)
plt.plot(x_new, y_interp, 'x')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Known data points
x_known = np.array([1, 2, 3, 4, 5])
y_known = np.array([2, 4, 3, 6, 8])

# Fit a linear model to the known data
coefficients = np.polyfit(x_known, y_known, 1)
slope, intercept = coefficients

# Define new x values for extrapolation
x_extrapolate = np.array([6, 7, 8])

# Perform linear extrapolation
y_extrapolate = slope * x_extrapolate + intercept

# Plotting
plt.plot(x_known, y_known, 'o', label='Known Data')
plt.plot(x_extrapolate, y_extrapolate, 'x', label='Extrapolated Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

print("Extrapolated Y values:", y_extrapolate)
```
