# Numerical Methods

## 1. Forward Difference ($\bigtriangleup$)

```python
y = np.random.randint(1,20,10)
```

```python
y
```

```python
def forward(A,n, ret):
    inner = []
    for i in range(len(A)-1):
        dy = A[i]-A[i+1]
        inner.append(dy)

    ret.append(inner)
    n = n-1
    if n>0:
        forward(inner,n,ret)
        
    return ret
```

```python
ret = []
forward(y,len(y)-1, ret)
```

```python
def f(x):
    return x**3 + 2*x + 1
```

```python
def fdiff(x,h):
    return ((f(x + h) - f(x)))
```

```python
y = np.array([1,2,3,4,5])
fdiff(y,1)
```

```python
for i in y:
    print(f"x = {i} \t f(x) = {f(i)} \t fdiff = {fdiff(i,1)}")
```
