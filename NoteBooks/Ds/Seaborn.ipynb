```python
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
%matplotlib inline
```

```python
from numpy.random import randn
```

# Numerical Data Plotting

```python
df = sns.load_dataset('tips')
```

```python
df.head(5)
```

```python
df.tail(5)
```

## RelPlot

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip')
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip', hue = 'smoker')
```

```python
df['smoker'].value_counts()
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip', hue = 'smoker', style = 'time')
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip', hue = 'size')
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip', hue = 'size', palette = 'ch:r=-0.5, l=0.75')
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip', size = 'size')
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip', size = 'size', sizes = (15,200))
```

## LinePlot

```python
df = pd.DataFrame(dict(time = np.arange(500), value = randn(500).cumsum()))
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
sns.relplot(data = df, x = 'time', y = 'value', kind = 'line')
```

```python
df = pd.DataFrame(dict(time = pd.date_range('2019-06-02', periods = 500), value = randn(500).cumsum()))
```

```python
df.head(5)
```

```python
f = sns.relplot(data = df, kind = 'line', x = 'time', y = 'value')
f.fig.autofmt_xdate()
```
