# Binomial Distribution

The binomial distribution is a probability distribution that models the number of successes in a fixed number of independent Bernoulli trials. In simpler terms, it describes the probability of getting a certain number of successes (usually denoted as "$k$") in a fixed number of attempts (usually denoted as "$n$"), where each attempt has two possible outcomes: *success* or *failure*.

* The key characteristics of a binomial distribution are as follows:

1. Independent Trials: Each trial is independent, meaning the outcome of one trial does not affect the outcome of another trial. For example, flipping a coin multiple times, each flip is independent of the others.

2. Two Outcomes: Each trial has only two possible outcomes, typically referred to as "success" and "failure." These outcomes are mutually exclusive, and the probability of success on a single trial is denoted by "p," while the probability of failure is denoted by "q," where q = 1 - p.

3. The probability mass function (PMF) of the binomial distribution is given by the formula:

$$P(\mathcal {X} = k) = C(n, k) \times p^k \times q^{(n-k)}$$

Where:

$P(\mathcal {X} = k)$ is the probability of getting exactly k successes in n trials.
C(n, k) is the binomial coefficient (also known as n choose k), representing the number of ways to choose k successes from n trials and is given by $C(n, k) = \frac{n!}{(k! \times (n-k)!)}$.
p is the probability of success on a single trial.
q = 1 - p is the probability of failure on a single trial.
The mean (expected value) and variance of the binomial distribution are given by:

$$Mean (μ) = n \times p$$
$$Variance (σ^2) = n \times p \times q$$

It is particularly useful when dealing with binary outcomes and counting the number of successes in a fixed number of trials. Some practical examples of situations modeled by the binomial distribution include:

- The number of defective items in a production batch.
- The number of customers who purchase a product out of a fixed number of potential customers.
- The number of heads obtained when flipping a coin multiple times.
- The binomial distribution is a fundamental concept in probability and forms the basis for more complex distributions and statistical methods in data science and related fields.

```python
n = 20 
p = 0.41
x = range(n + 1)

pmf_values = stat.binom.pmf(x, n, p)

plt.bar(x, pmf_values)
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.show()
```

```python
mean, var, skew, kurt = stat.binom.stats(n, p, moments='mvsk')
```

```python
mean, var, skew, kurt
```

```python
stat.binom.cdf(2, n, p, loc=0) # Cumulative distribution function
```

# Poisson binomial distribution

The Poisson binomial distribution is a probability distribution that generalizes the binomial distribution to scenarios where the individual trials have different success probabilities. In other words, it describes the number of successes in a sequence of independent trials, where each trial can have its own probability of success.

In the classic binomial distribution, each trial has the same probability of success (denoted as "$p$"). However, in the Poisson binomial distribution, the probability of success can vary from trial to trial. Let's say we have a sequence of n independent trials, and the probability of success for each trial is given by an array of probabilities: $p = [p1, p2, ..., pn]$, where $p_i$ represents the probability of success for the ith trial.

The probability mass function (PMF) of the Poisson binomial distribution is more complex than that of the binomial distribution. It involves calculating the probability of each possible number of successes (k) in the n trials, taking into account the different success probabilities for each trial.

The PMF of the Poisson binomial distribution is given by the formula:

$$P(X = k) = ∑[(-1)^r * C(n, r) * e^{(i*pi)} * (1 - pi)^{(n-r)}], for k = 0, 1, 2, ..., n$$

Where:

$P(X = k)$ is the probability of getting exactly k successes in n trials.
$C(n, r)$ is the binomial coefficient (n choose r) representing the number of ways to choose r successes from n trials and is given by $C(n, r) = \frac{n!}{(r! \times (n-r)!)}$.
$e$ is the base of the natural logarithm, approximately equal to 2.71828.
$i$ is the imaginary unit $(i^2 = -1)$.
pi represents the probability of success for the ith trial $(0 ≤ \pi ≤ 1)$.
The summation is taken over all subsets of trials that result in r successes.
Calculating the PMF directly for large values of n can be computationally intensive, but there are efficient algorithms and methods available to approximate the Poisson binomial distribution.

The Poisson binomial distribution is particularly useful when dealing with scenarios where individual trials have different success probabilities, such as in finance, epidemiology, and other real-world applications. It provides a flexible and general framework to model situations where the assumption of constant success probability across trials is not valid.

```python
mu = 0.6
mean, var, skew, kurt = stat.poisson.stats(mu, moments='mvsk')
```

```python
mean, var, skew, kurt
```

```python
x = range(10)

pmf = stat.poisson.pmf(x, mu)

plt.bar(x, pmf)
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title(f"Poisson Binomial Distribution (mu = {mu})")
plt.show()
```
