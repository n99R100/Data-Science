# Area under the  Lines

```python
f = 2*x
f
```

```python
x_values = np.linspace(-10,10,1000)
```

```python
fig = plt.figure(figsize = (10,3), layout = 'tight')

gs = GridSpec(1,1, figure = fig)

ax = fig.add_subplot(gs[0])

ax.plot(x_values, [f.subs(x,i) for i in x_values], label = r'f(x)')

ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.minorticks_on()

plt.show()
```

- Area between $0 \leq x \leq 5$

```python
y = np.array([sp.N(f.subs(x,i)) for i in x_values])
```

```python
fig = plt.figure(figsize = (10,3), layout = 'tight')

gs = GridSpec(1,1, figure = fig)

ax = fig.add_subplot(gs[0])

ax.plot(x_values, y, label = r'f(x)')

f_numpy = sp.lambdify(x, f, modules='numpy')
y_values = f_numpy(x_values)

plt.fill_between(x_values, 0, y_values, where=(0 <= x_values) & (x_values <= 5), alpha=0.2, 
                 color='blue', label=r'Shaded Area $(0 \leq x \leq 5)$')

ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.minorticks_on()

plt.show()
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sp.integrate(f,(x,0,5))
```

- Area between $-5 \leq x \leq 5$

```python
fig = plt.figure(figsize = (10,3), layout = 'tight')

gs = GridSpec(1,1, figure = fig)

ax = fig.add_subplot(gs[0])

ax.plot(x_values, y, label = r'f(x)')

f_numpy = sp.lambdify(x, f, modules='numpy')
y_values = f_numpy(x_values)

plt.fill_between(x_values, 0, y_values, where=(-5 <= x_values) & (x_values <= 5), alpha=0.2, 
                 color='blue', label=r'Shaded Area $(-5 \leq x \leq 5)$')

ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.minorticks_on()

plt.show()
```

```python
sp.integrate(f,(x,0,5)) - sp.integrate(f,(x,-5,0))
```

## Between two lines

```python
f = 2*x+3
g = -2*x-3
```

```python
fig = plt.figure(figsize = (10,3), layout = 'tight')

gs = GridSpec(1,1, figure = fig)

ax = fig.add_subplot(gs[0])

ax.plot(x_values, [f.subs(x,i) for i in x_values], label = r'f(x)')
ax.plot(x_values, [g.subs(x,i) for i in x_values], label = r'g(x)')

ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.minorticks_on()

plt.show()
```

- Area between $0 \leq x \leq 7$

```python
fig = plt.figure(figsize = (10,3), layout = 'tight')

gs = GridSpec(1,1, figure = fig)

ax = fig.add_subplot(gs[0])

ax.plot(x_values, [f.subs(x,i) for i in x_values], label = r'f(x)')
ax.plot(x_values, [g.subs(x,i) for i in x_values], label = r'g(x)')
ax.fill_between(x_values, np.array([f.subs(x,i) for i in x_values], dtype = float),
                np.array([g.subs(x,i) for i in x_values], dtype = float),
               where = (0 <= x_values)&(x_values <= 7), color='red', alpha = 0.2, label=r'Shaded Area $(0 \leq x \leq 7)$')

ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.minorticks_on()

plt.show()
```

```python
sp.integrate(f,(x,0,7))-sp.integrate(g,(x,0,7))
```

# Area under the curve

```python
f = x**2 + 2*x -1
f
```

- Area between $0.7 \leq x \leq 3.7$

```python
x_values =  np.linspace(-5,5,1000)
```

```python
fig = plt.figure(figsize = (7,4), layout = 'tight')

gs = GridSpec(2,2, figure = fig)

ax1 = fig.add_subplot(gs[:,0])
ax2 = fig.add_subplot(gs[:,1])

ax1.plot(x_values, [f.subs(x,i) for i in x_values], label = r'f(x)')
ax2.plot(x_values, [f.subs(x,i) for i in x_values], label = r'f(x)')

ax2.fill_between(x_values, 0, np.array([f.subs(x,i) for i in x_values], dtype = float), 
                 where=(0.7 <= x_values) & (x_values <= 3.7), alpha=0.2, 
                 color='navy', label=r'Shaded Area $(0.7 \leq x \leq 3.7)$')

ax1.legend()
ax2.legend()
fig.supxlabel('x')
fig.supylabel('y')
ax1.minorticks_on()

plt.show()
```

```python
sp.integrate(f,(x,0.7,3.7))
```

## Between two curves

```python
f = 2*x**2-2*x-2
f
```

```python
g = x**3+x**2-12*x
g
```

```python
x_values =  np.linspace(-5,5,1000)
```

```python
f_val = np.array([f.subs(x,i) for i in x_values], dtype = float)
g_val = np.array([g.subs(x,i) for i in x_values], dtype = float)
```

```python
fig = plt.figure(figsize = (10,4), layout = 'tight')

gs = GridSpec(2,3, figure = fig)

ax1 = fig.add_subplot(gs[:-1,0])
ax2 = fig.add_subplot(gs[-1,0])
ax3 = fig.add_subplot(gs[:,1])
ax4 = fig.add_subplot(gs[:,2])

ax1.plot(x_values, f_val, label = r'f(x)')
ax2.plot(x_values, g_val, label = r'g(x)')


ax3.fill_between(x_values, f_val, g_val, where = (-2.81082102737248 <= x_values)&(x_values <= 0.196886779471732), 
                 color='navy', alpha = 0.2)

ax4.fill_between(x_values, f_val, g_val, where = (0.196886779471732 <= x_values)&(x_values <= 3.61393424790075), 
                 color='blue', alpha = 0.2)

for i in [ax1,ax2]:
    i.set_xlabel('x')
    i.set_ylabel('y')
    if (i.get_legend_handles_labels()):
        i.set_title((i.get_legend_handles_labels()[1])[0])

plt.show()
```

```python
sp.integrate(g,(x,-2.81082102737248,0.196886779471732)) - sp.integrate(f,(x,-2.81082102737248,0.196886779471732))
```

```python
sp.integrate(f, (x,0.196886779471732,3.61393424790075)) - sp.integrate(g, (x,0.196886779471732,3.61393424790075))
```

## Between curve and a line

```python
f = x**2-6
g = -2*x+2
```

```python
f_val = np.array([f.subs(x,i) for i in x_values], dtype = float)
g_val = np.array([g.subs(x,i) for i in x_values], dtype = float)
```

```python
fig = plt.figure(figsize = (10,4), layout = 'tight')

gs = GridSpec(2,3, figure = fig)

ax1 = fig.add_subplot(gs[:,0])
ax2 = fig.add_subplot(gs[:,1:])

ax1.plot(x_values, f_val, label = r'f(x)')
ax1.plot(x_values, g_val, label = r'g(x)')


ax2.fill_between(x_values, f_val, g_val, where = (-4 <= x_values)&(x_values <= 2), 
                 color='navy', alpha = 0.2)

for i in [ax1,ax2]:
    i.set_xlabel('x')
    i.set_ylabel('y')

    ax1.legend()

plt.show()
```

```python
sp.integrate(g,(x,-4,1)) - sp.integrate(f,(x,-4,1))
```
