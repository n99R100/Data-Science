# Single Layout

```python
x = np.linspace(-np.pi/2, np.pi/2,1000)
y = np.sin(x)
```

## Plot functions and their solutions

### 1. Find solutions

```python
x = sym.symbols('x', real = True)
```

```python
f = 2*sym.exp(x/5)*sym.sin(x)
f
```

```python
g = 2*sym.exp(x/5)*sym.cos(x)
g
```

#### Zeros:

```python
fz = sym.solveset(f,x, sym.Interval(-2*np.pi, 2*np.pi))
fz
```

```python
gz = sym.solveset(g,x, sym.Interval(-2*np.pi, 2*np.pi))
gz
```

#### Crosses:

```python
fg = sym.solveset(sym.Eq(f,g),x, sym.Interval(-2*np.pi, 2*np.pi))
fg
```

### 2. Plot results

```python
x_values = np.linspace(-2*np.pi, 2*np.pi,1000)
```

```python
fig = plt.figure( figsize = (10,6), layout = 'constrained')

sf = fig.subfigures(1, 2, wspace=0.07, width_ratios=[2, 1])

gs = GridSpec(2,2, figure = sf[0])

ax1 = sf[0].add_subplot(gs[0,0])
ax2 = sf[0].add_subplot(gs[0,1])
ax3 = sf[0].add_subplot(gs[1,:])

ax4,ax5 = sf[1].subplots(2,1)

ax1.plot(x_values,[f.subs(x,val) for val in x_values], label = r'$y_1$')
ax1.minorticks_on()
ax1.set_title(r'$y_1$')
ax1.set_xlabel(r'$\theta$')
ax1.set_ylabel(r'$f(\theta)$')
ax1.legend(loc ='best')

ax2.plot(x_values,[g.subs(x,val) for val in x_values], label = r'$y_2$')
ax2.minorticks_on()
ax2.set_title(r'$y_2$')
ax2.set_xlabel(r'$\theta$')
ax2.set_ylabel(r'$g(\theta)$')
ax2.legend(loc ='best')

ax3.plot(x_values,[f.subs(x,val) for val in x_values], label = r'$y_1$')
ax3.plot(x_values,[g.subs(x,val) for val in x_values], label = r'$y_2$')
ax3.scatter([float(i) for i in fz],[0]*len(fz), label = r'$y_1 = 0$')
ax3.scatter([float(i) for i in gz],[0]*len(gz), label = r'$y_2 = 0$')
ax3.minorticks_on()
ax3.set_title(r'$y_1$ and $y_2$')
ax3.legend(loc ='upper left')
ax3.spines[['left','bottom']].set_position('zero')
ax3.spines[['right', 'top']].set_visible(False)

ax4.scatter([float(i) for i in fz],[0]*len(fz), label = r'$y_1 = 0$')
ax4.scatter([float(i) for i in gz],[0]*len(gz), label = r'$y_2 = 0$')
ax4.minorticks_on()
ax4.set_title(r'Zeros of $y_1$ and $y_2$')
ax4.legend(loc ='best')
ax4.spines[['left','bottom']].set_position('zero')
ax4.spines[['right', 'top']].set_visible(False)

ax5.scatter([float(i) for i in fg],[f.subs(x,val) for val in fg], label = r'$y_1 = y_2$')
ax5.minorticks_on()
ax5.set_title(r'Cross of $y_1$ and $y_2$')
ax5.legend(loc ='best')
ax5.spines[['left','bottom']].set_position('zero')
ax5.spines[['right', 'top']].set_visible(False)

fig.suptitle('Plots and Solutions', fontsize = 25, color = 'blue')
sf[0].suptitle('Plots')
sf[1].suptitle('Solutions')

plt.show()
```
