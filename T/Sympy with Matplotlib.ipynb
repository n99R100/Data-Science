```python
import numpy as np
import sympy as sp
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
sns.set_theme(context='notebook', style='ticks', palette='pastel', 
              font='sans-serif', font_scale=1, color_codes=True, rc=None)
```

```python
x = sp.symbols('x', real = True)
```

# Plot functions

## Single function

```python
f = x**2
f
```

```python
x_values = np.linspace(-5,5,1000)
```

```python
fig = plt.figure(figsize = (5,5), layout = 'tight')

plt.plot(x_values, [f.subs(x,i) for i in x_values])

plt.show()
```

## Multiple functions

```python
f = sp.sin(x)
f
```

```python
g = sp.cos(x)
g
```

```python
x_values = np.linspace(-2*np.pi, 2*np.pi, 1000)
```

```python
fig = plt.figure(figsize = (10,3), layout = 'tight')

plt.plot(x_values, [f.subs(x,i) for i in x_values])
plt.plot(x_values, [g.subs(x,i) for i in x_values])

plt.show()
```

# Plot functions with solutions

```python
f = 2*sp.exp(x/8)*sp.sin(3*x)
f
```

```python
sol = sp.solveset(f,x,sp.Interval(-2*np.pi, 2*np.pi))
sol
```

```python
x_values = np.linspace(-2*np.pi, 2*np.pi, 1000)
```

```python
fig = plt.figure(figsize = (10,3), layout = 'constrained')

plt.plot(x_values, [f.subs(x,i) for i in x_values])
plt.scatter([float(i) for i in sol], [f.subs(x,i) for i in sol])

plt.show()
```

```python
f = x**2+2*x-5
f
```

```python
g = 2*x**2-5*x-6
g
```

```python
solf = sp.solveset(f,x,sp.Interval(-10,10))
solf
```

```python
solg = sp.solveset(g,x,sp.Interval(-10,10))
solg
```

```python
solfg = sp.solve(sp.Eq(f,g),x)
solfg
```

```python
x_values = np.linspace(-10, 10, 1000)
```

```python
fig = plt.figure(figsize = (10,3), layout = 'constrained')

gs = GridSpec(1,2, width_ratios=[2,1], figure = fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])

ax1.plot(x_values, [f.subs(x,i) for i in x_values], label = r'$y_1$')
ax1.plot(x_values, [g.subs(x,i) for i in x_values], label = r'$y_2$')

ax2.scatter([float(i) for i in solf], [f.subs(x,i).n() for i in solf], label = r'$y_1 = 0$')
ax2.scatter([float(i) for i in solg], [g.subs(x,i).n() for i in solg], label = r'$y_2 = 0$')
ax2.scatter([float(i.n()) for i in solfg], [g.subs(x,i).n() for i in solfg], label = r'$y_1 = y_2$')

ax1.legend()
ax2.legend(loc = 'center right')

ax2.spines[['left','bottom']].set_position('zero')
ax2.spines[['right', 'top']].set_visible(False)

plt.show()
```

```python
f = 2*sp.log(x)*sp.sin(x+2)
f
```

```python
sol = sp.solveset(f,x, sp.Interval(1,10))
sol.n()
```

```python
x_values = np.linspace(0.1,10,1000)
```

```python
fig = plt.figure(figsize = (10,3), layout = 'tight')

plt.plot(x_values, [f.subs(x,i) for i in x_values])
plt.scatter([float(i) for i in sol],[f.subs(x,i) for i in sol.n()])

plt.show()
```
