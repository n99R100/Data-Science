# Asymtote

<div class="alert alert-block alert-info">
    <b><font color = 'blue'; size = 4> Definition:</font></b> <hr>
An asymptote is a straight line or curve that a mathematical function or curve approaches but never quite reaches. In other words, it's a line that the function gets arbitrarily close to as the independent variable (usually denoted as "$x$") moves to infinity ($+\infty$) or negative infinity ($-\infty$).
</div>

# Linear denominator

# 1. $f:x \mapsto \frac{1}{x}$

<div class="alert alert-block alert-warning">
<b>Note:</b>
<hr>
$f(x)= \frac{1}{x}$ has a horizontal asymptote at $y=0$ and a vertical asymptote at $x=0$.
</div>

```python
f = 1/x
f
```

```python
x_l = np.linspace(-5,-0.05,1000)
x_r = np.linspace(0.05,5,1000)

y_l = np.array([sp.N(f.subs(x,i)) for i in x_l])
y_r = np.array([sp.N(f.subs(x,i)) for i in x_r])
```

```python
fig = plt.figure(figsize = (7,5), layout = 'tight')

gs = GridSpec(1,1, figure = fig)

ax = fig.add_subplot(gs[0,0])

ax.plot(x_l,y_l)
ax.plot(x_r,y_r)
ax.axvline(x=0,color='.5', linestyle='--', label=r'$x=0$')
ax.axhline(y=0,color='.7', linestyle='--', label=r'$y=0$')

ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(r'$f:x \mapsto \frac{1}{x}$')
ax.minorticks_on()

fig.suptitle('Rational Functions')
plt.show()
```

# 2. $f:x \mapsto \frac{3x^2 +2x -1}{x+2}$

```python
f = (3*x**2+2*x-1)/(x+2)
f
```

```python
f.apart()
```

<div class="alert alert-block alert-warning">
<b>Note:</b>
<hr>
$f(x)=  \frac{3 x^{2} + 2 x - 1}{x + 2}$ has a horizontal asymptote at $y=3x-4$ and a vertical asymptote at $x=-2$.
</div>

```python
x_l = np.linspace(-10,-2.1,1000)
x_r = np.linspace(-1.9,10,1000)

y_l = np.array([sp.N(f.subs(x,i)) for i in x_l])
y_r = np.array([sp.N(f.subs(x,i)) for i in x_r])
```

```python
fig = plt.figure(figsize = (7,5), layout = 'tight')

gs = GridSpec(1,1, figure = fig)

ax = fig.add_subplot(gs[0,0])

ax.plot(x_l,y_l)
ax.plot(x_r,y_r)
plt.axline((0, -4), (10,26), color="red", linestyle=(0, (5, 5)))
ax.axvline(x=-2,color='green', linestyle='--', label=r'$x=0$')

#ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(r'$f:x \mapsto \frac{3x^2 +2x -1}{x+2}$')
ax.minorticks_on()
ax.grid()

fig.suptitle('Rational Functions')
plt.show()
```

# 3. $f:x \mapsto \frac{x^{2} + 5 x + 6}{x - 3}$

```python
f = (x**2+5*x+6)/(x-3)
f
```

```python
f.apart()
```

<div class="alert alert-block alert-warning">
<b>Note:</b>
<hr>
$f(x)=  \frac{x^{2} + 5 x + 6}{x - 3}$ has a horizontal asymptote at $y=x+8$ and a vertical asymptote at $x=3$.
</div>

```python
x_l = np.linspace(-3,2.9,1000)
x_r = np.linspace(3.1,9,1000)

y_l = np.array([sp.N(f.subs(x,i)) for i in x_l])
y_r = np.array([sp.N(f.subs(x,i)) for i in x_r])
```

```python
fig = plt.figure(figsize = (7,5), layout = 'tight')

gs = GridSpec(1,1, figure = fig)

ax = fig.add_subplot(gs[0,0])

ax.plot(x_l,y_l)
ax.plot(x_r,y_r)
plt.axline((0, 8), (10,18), color="red", linestyle=(0, (5, 5)))
ax.axvline(x=3,color='green', linestyle='--', label=r'$x=0$')

#ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(r'$f:x \mapsto \frac{x^{2} + 5 x + 6}{x - 3}$')
ax.minorticks_on()
ax.grid()

fig.suptitle('Rational Functions')
plt.show()
```

# Non-Linear Denominator

# 1. $f:x \mapsto \frac{5 x^{2}}{4 x^{2} + 8 x - 5}$

```python
f = (5*x**2)/(4*x**2+8*x-5)
f
```

```python
f.apart()
```

```python
x_l1 = np.linspace(-5,-2.6,1000)
x_r1 = np.linspace(-2.4,0,1000)

x_l2 = np.linspace(0,0.49,1000)
x_r2 = np.linspace(0.51,2,1000)

y_l1 = np.array([sp.N(f.subs(x,i)) for i in x_l1])
y_r1 = np.array([sp.N(f.subs(x,i)) for i in x_r1])

y_l2 = np.array([sp.N(f.subs(x,i)) for i in x_l2])
y_r2 = np.array([sp.N(f.subs(x,i)) for i in x_r2])
```

```python
fig = plt.figure(figsize = (7,5), layout = 'tight')

gs = GridSpec(1,1, figure = fig)

ax = fig.add_subplot(gs[0,0])

ax.plot(x_l1,y_l1)
ax.plot(x_r1,y_r1)
ax.plot(x_l2,y_l2)
ax.plot(x_r2,y_r2)
#plt.axline((0, 8), (10,18), color="red", linestyle=(0, (5, 5)))
ax.axvline(x=-2.5,color='red', linestyle='--', label=r'$x=0$')
ax.axvline(x=0.5,color='red', linestyle='--', label=r'$x=0$')

#ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(r'$f:x \mapsto \frac{5 x^{2}}{4 x^{2} + 8 x - 5}$')
ax.minorticks_on()
ax.grid()

fig.suptitle('Rational Functions')
plt.show()
```

# 2. $f:x \mapsto \frac{\left(x - 4\right) \left(3 x - 2\right)}{\left(x - 1\right) \left(x + 4\right)}$

```python
f = ((3*x-2)*(x-4))/((x-1)*(x+4))
f
```

```python
f.apart()
```

```python
sp.plot((f,(x,-6,-4.1)),(f,(x,-3.9,0)),(f,(x,0,0.997)),(f,(x,1.002,5)), 
        title = r'$f:x \mapsto \frac{\left(x - 4\right) \left(3 x - 2\right)}{\left(x - 1\right) \left(x + 4\right)}$',
       xlabel = None, ylabel = None)
plt.show()
```

# 3. $f:x \mapsto \frac{\left(x + 4\right) \left(2 x + 5\right)}{x^{2} + x + 5}$

```python
f = ((2*x+5)*(x+4))/(x**2+x+5)
f
```

```python
f.apart()
```

```python
x_vals = np.linspace(-100,100,1000)
y_vals = np.array([sp.N(f.subs(x,i)) for i in x_vals])
```

```python
fig = plt.figure(figsize = (7,5), layout = 'tight')

gs = GridSpec(1,1, figure = fig)

ax = fig.add_subplot(gs[0,0])

ax.plot(x_vals,y_vals)

ax.axhline(y=2,color='red', linestyle='--', label=r'$y = 2$')

ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(r'$f:x \mapsto \frac{\left(x + 4\right) \left(2 x + 5\right)}{x^{2} + x + 5}$')
ax.minorticks_on()
ax.grid()

fig.suptitle('Rational Functions')
plt.show()
```
