# Packages

```python
import numpy as np
import sympy as sp
```

```python
import matplotlib as mat
import mpl_toolkits as mpl
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.gridspec import GridSpecFromSubplotSpec
```

```python
x = np.linspace(-10,10,20)
y = x
```

# Line Plot

```python
plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure()

plt.title("Plot of y = x")
plt.plot(x, x, 'red',label = "y = x", linewidth = 5)
plt.plot(x, 2*x, 'blue',label = "y = 2x")
plt.plot(x, x**2, 'green',label = r"$y = x^2$")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.show()
```

# Scatter Plot

```python
fig = plt.figure()

plt.title(r"Plot of $y = x$")
plt.scatter(x, x**2,color = 'red', alpha = 0.5,label = r"$y = x^2$")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.show()
```

# Subplots

```python
x = np.linspace(-5,5,1000)
y1 = x**2-2*x-5
y2 = 2*x**3 - 4*x**2 + 5*x + 3
```

```python
fig = plt.figure(figsize = (10,5), layout = 'constrained')

gs = GridSpec(1 ,2 , figure = fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])

ax1.plot(x,y1, label = r'$y = x^2 -2x -5$')
ax1.set_title('Squared Plot')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax1.grid()

ax2.plot(x,y2, color = 'red', label = r'$y = 2x^3 - 4x^2 + 5x + 3$')
ax2.set_title('Cubic Plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()
ax2.grid()

plt.suptitle('Sub plots', fontsize = 20)

plt.show()
```

```python
fig = plt.figure( figsize = (10,7), layout = "constrained")

gs = GridSpec(2, 2, figure = fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.plot(x,x**2, label = r'$x^2$')
ax2.plot(x,x**3, color = 'red',label = r'$x^3$')
ax3.plot(x,x**2-2*x-3, color = 'green',label = r'$x^2 -2x -3$')
ax4.plot(x,x**3 -3*x**2 + 4*x +6, color = 'navy',label = r'$x^3 -3x^2 +4x +6$')

t = [r'$x^2$', r'$x^3$', r'$x^2 -2x -3$', r'$x^3 -3x^2 +4x +6$']
ax = [ax1,ax2,ax3,ax4]

for i,j in zip(ax,t):
    i.set_title(j)
    i.legend()
    i.grid()
    i.set_xlabel('x')
    i.set_ylabel('y')

plt.suptitle('Multiple Plots', fontsize = 20)
    
plt.show()
```

```python
fig = plt.figure(figsize = (10,7), layout = 'constrained')

gs = GridSpec(1,2, figure = fig)

left = GridSpecFromSubplotSpec(3, 3, subplot_spec = gs[0])
right = GridSpecFromSubplotSpec(3, 3, subplot_spec = gs[1])

ax1 = fig.add_subplot(left[:-1,:])
ax2 = fig.add_subplot(left[-1,:-1])
ax3 = fig.add_subplot(left[-1,-1])
ax4 = fig.add_subplot(right[:,:-1])
ax5 = fig.add_subplot(right[:-1,-1])
ax6 = fig.add_subplot(right[-1,-1])

ax = [ax1,ax2,ax3,ax4,ax5,ax6]
y = [x,2*x,x**2,3*x+1,3*x**2-1,np.sin(x)]

for j,i,k in zip(range(6),ax,y):
    #i.legend()
    i.grid()
    i.set_xlabel('x')
    i.set_ylabel('y')
    i.set_title(str(j))
    i.plot(x,k)
    i.spines[['left', 'bottom']].set_position('zero')
    i.spines[['right', 'top']].set_visible(False)

fig.suptitle('Plot', fontsize = 20)    

plt.show()
```

```python
fig, ax = plt.subplots(1,1)


ax.set_title("Plot of y = x")
ax.plot(x, x, 'red',label = "y = x")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.spines[['left', 'bottom']].set_position('zero')
ax.spines[['right', 'top']].set_visible(False)

ax.grid()

plt.show()
```

```python
fig, ax = plt.subplots(1,1)


ax.set_title("Plot of y = x")
ax.plot(x, x**2, 'red',label = "y = x")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.spines[['left', 'bottom']].set_position('zero')
ax.spines[['right', 'top']].set_visible(False)
ax.fill_between(x,x**2,  where = ((2 < x) & (x <4 )), alpha = 0.2)
ax.fill_betweenx(x**2,x,  where = ((-4 < x) & (x < -2 )), alpha = 0.2)

ax.grid()

plt.show()
```

```python
fig = plt.figure( figsize = (10,4), layout = 'constrained')

sf = fig.subfigures(1, 2, wspace=0.07, width_ratios=[2, 1])

left = sf[0].subplots(1,2, sharey = True)
right = sf[1].subplots(3,1, sharex = True)

sf[0].suptitle('Left')
sf[1].suptitle('Right')

fig.suptitle('Super Figure')

left[0].set_xlabel('x')
left[1].set_xlabel('x')
right[0].set_xlabel('x')
right[1].set_xlabel('x')
right[2].set_xlabel('x')

left[0].set_ylabel('y')
left[1].set_ylabel('y')
right[0].set_ylabel('y')
right[1].set_ylabel('y')
right[2].set_ylabel('y')

sf[0].supxlabel('X Values')
sf[1].supxlabel('X Values')
sf[0].supylabel('Y Values')
sf[1].supylabel('Y Values')

sf[0].set_facecolor('0.95')
sf[1].set_facecolor('0.95')

plt.show()
```
