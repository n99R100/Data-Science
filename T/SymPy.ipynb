## Solve equations

```python
n = sym.symbols('n')
```

```python
eq = sym.Eq(n+3,11)
eq
```

```python
sym.solve(eq,n)
```

```python
sym.solve(sym.Eq(2*n**2+34,2*n**2+n),n)
```

## Simulteneous equations

```python
eq1 = sym.Eq(5*x+3*y,19)
eq2 = sym.Eq(x+3*y, 11)
```

```python
eq1
```

```python
eq2
```

```python
res = sym.solve([eq1,eq2],x,y)
res
```

```python
eq1.subs({x:res[x], y:res[y]})
```

```python
eq = sym.Eq(x**2+6*x+6,0)
eq
```

```python
res = sym.solve(eq,x)
res
```

```python
res[0]
```

```python
res[1]
```

## Lambdify

```python
f = ((17*x-53)/(x**2-2*x-15)).apart()
f
```

```python
h = sym.lambdify(x, f, 'numpy')
h
```

```python
h(4)
```

```python
f = (sym.sin(x + sym.pi/8)*sym.cos(2*x+sym.pi/4))/(2*sym.sin(x) + sym.Rational(1,3)*sym.cos(-x))
f
```

```python
f.subs(x,sym.pi).evalf(15)
```

```python
h = sym.lambdify(x, f, 'numpy')
```

```python
h(np.pi)
```
